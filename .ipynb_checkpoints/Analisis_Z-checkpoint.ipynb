```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./QRO_DIP_LOC_2018.csv",skiprows=4)
df.head()
```

Como podemos observar dando un vistazo a nuestros datos y complementando con la información del catalogo (LEEME.txt), cada registro representa los votos totales por cada casilla electoral, obteniendo un total de 2648 de casillas para esta elección

Identifiquemos las columnas y algunos de sus atributos

```python
df.info()
```

Como podemos notar, una gran cantidad de columnas, incluyendo las que contabilizan los votos para cada partido (16-68), son de tipo "object", lo cual nos indica que pueden existen datos que varían en su tipo de dato  
Bajo esta observación, podemos justificar un posterior analisis de dichas columnas y en base a los descubrimientos podremos realizar una propuesta de limpieza de estos datos

Tambien veamos los valores no-numericos en los votos

```python
non_numeric_values = []
columns = df.columns
for i in range(16,69):
    uniques = df[columns[i]].unique()
    for y in uniques:
        #print(type(y),y)
        if type(y) is str and not y.isdigit():
            non_numeric_values.append(y)

set(non_numeric_values)
```

```python
df.head(50).iloc[20:40,16:69]
```

```python
# Del data frame, filtramos los valores que son nulos en cualquier columna de votos (este comportamiento aparece en todos los votos)
# Consultamos la columna "OBSERVACIONES" y contamos los valores que aparecen en ella
df[df["PAN"].isnull()]["OBSERVACIONES"].value_counts()
```

Como podemos ver, los datos no numericos tienen las siguientes explicaciones:
- \- : es usado en los apartados de candidatos independientes que no participan como candidatos en la zona de elección
- NaN: en el archivo original pueden verse como celdas en blanco. Este tipo de registros vienen acompañados por una explicación en la columna "explicación", las cuales se listan en la casilla anterior

Mientras que los apartados con Ilegible/Sin dato se explican por si mismos

Con esto en mente, podemos proponer como tratamiento eliminar los registros de casillas donde no se realizó la votación, y volver los demas valores cero

```python
for i in range(16,69):
    df[columns[i]] = df[columns[i]].replace({"Sin dato":"0","Ilegible":"0","-":"0"})
```

```python
df.head(50).iloc[20:40,16:69]
```

Los registros nulos no aportan información relevante al analisis de la información en general, pero pueden ser relevantes si se analizan por separado

```python
df_nulls = df[df["PAN"].isnull()]
df = df.dropna(subset=["PAN"])
```

```python
df_nulls.iloc[:5,16:69]
```

```python
df.iloc[:5,16:69]
```

Pero las columnas aún tienen el tipo de dato "object", lo cual limita la forma en que podemos trabajar con estas columnas, asi que las transformaremos a int

```python
for i in range(16,69):
    df[columns[i]]= df[columns[i]].fillna(0)
    df[columns[i]] = df[columns[i]].astype(int)
df.iloc[:5,16:69]
```

```python
df_filtered = df.iloc[:,16:69]
df_filtered = df_filtered.join(df["DISTRITO_LOCAL"])
```
